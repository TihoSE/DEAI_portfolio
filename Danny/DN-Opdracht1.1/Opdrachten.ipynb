## While-loop

Schrijf de inhoud van de methode wachtwoord. Deze heeft geen parameters en laat de gebruiker twee wachtwoorden invoeren, waarna hij bekijkt of deze gelijk zijn. Zolang dat niet het geval is wordt er eerst Fout, probeer het opnieuw! uitgeprint en moet de gebruiker het opnieuw proberen.<br>Als de gebruiker het goed doet, komt er Succes! op het scherm te staan.

```python
def password_checker():
    # Eerste check om te controleren of beide wachtwoorden met elkaar 
    wachtwoord = input("Vul uw wachtwoord in")
    herhaal_wachtwoord = input("Herhaal uw wachtwoord")

    #  Als de wachtwoorden niet overeenkomt, dan moet de gebruiker het nog een keer proberen.
    while wachtwoord != herhaal_wachtwoord:
        print("Fout, probeer het opnieuw!")
        wachtwoord = input("Vul uw wachtwoord in")
        herhaal_wachtwoord = input("Herhaal uw wachtwoord")

    print("Succes!")

# Functie oproepen
password_checker()
```

## Objectoriëntatie

Maak een minisysteem voor een kledingwinkel. Maak verschillende klassen voor verschillende soorten kledingstukken, waarbij sommige klassen geërfd worden van andere klassen.
- Maak een klasse Kledingstuk aan, waarin door de constructor een merk, een maat en een prijs geïnitialiseerd worden.
- Maak een klasse Broek en een klasse Shirt, beide erven van de klasse Kledingstuk. Voeg aan Broek een broek_type, en aan Shirt een shirt_type toe (d.m.v. constructorgebruik).
- Maak een klasse Spijkerbroek die erft van broek en een klasse Poloshirt die erft van shirt. Voeg aan Spijkerbroek een blauwtint toe, en aan Poloshirt een kraagsoort (d.m.v. constructorgebruik).
- Implementeer in elke klasse een beschrijvingmethode, waarmee alle informatie over het betreffende kledingstuk wordt afgedrukt.
- Maak van elke klasse een object aan en roep de beschrijvingmethode aan. Controleer nauwkeurig of alles naar behoren werkt.

```python
class Kledingstuk:
    merk: str
    maat: float
    prijs: float

    def __init__(self, merk: str, maat: float, prijs: float):
        self.merk = merk
        self.maat = maat
        self.prijs = prijs

    def beschrijving(self):
        return(f"Informatie voor een kledingstuk (Parent klasse):\nMerk: {self.merk}\nMaat: {self.maat}\nPrijs: {self.prijs}\n")

# ======Erft van de klasse Kledingstuk ===========
class Broek(Kledingstuk):
    broek_type: str

    def __init__(self, merk: str, maat: float, prijs: float, broek_type: str):
        super().__init__(merk, maat, prijs) 
        self.broek_type = broek_type
    
    def beschrijving(self):
        return(f"Informatie voor een broek (child klasse):\nMerk: {self.merk}\nMaat: {self.maat}\nPrijs: {self.prijs}\nBroek type: {self.broek_type}\n")


class Shirt(Kledingstuk):
    shirt_type: str

    def __init__(self, merk: str, maat: float, prijs: float, shirt_type: str):
        super().__init__(merk, maat, prijs) #Called from the Kledingstuk Class :)
        self.shirt_type = shirt_type

    def beschrijving(self):
        return(f"Informatie voor een broek (child klasse):\nMerk: {self.merk}\nMaat: {self.maat}\nPrijs: {self.prijs}\nShirt type: {self.shirt_type}\n")

# =======Erft een van de klasses Broek en Shirt ============
class Spijkerbroek(Broek):
    blauwetint: str

    def __init__(self, merk: str, maat: float, prijs: float, broek_type: str, blauwetint: str):
        super().__init__(merk, maat, prijs, broek_type)
        self.blauwetint = blauwetint

    def beschrijving(self):
        return(f"Informatie voor een broek (grandchild klasse):\nMerk: {self.merk}\nMaat: {self.maat}\nPrijs: {self.prijs}\nBroek type: {self.broek_type}\nBlauwetint: {self.blauwetint}\n")

class Poloshirt(Shirt):
    kraagsoort: str

    def __init__(self, merk: str, maat: float, prijs: float, shirt_type: str, kraagsoort: str):
        super().__init__(merk, maat, prijs, shirt_type)
        self.kraagsoort = kraagsoort

    def beschrijving(self):
        return(f"Informatie voor een broek (grandchild klasse):\nMerk: {self.merk}\nMaat: {self.maat}\nPrijs: {self.prijs}\nShirt type: {self.shirt_type}\nKraagsoort: {self.kraagsoort}\n")
        

# ============Buiten de klasses ==============

KledingTest1 = Kledingstuk("KledingStuk van 6 7", 67.67, 67.67)
KledingTest2 = Broek("Broek van Minji", 169.00, 16.90, "NewJeans")
KledingTest3 = Shirt("Shirt van Sonic", 67.67, 67.67, "SEGA")
KledingTest4 = Spijkerbroek("Spijkerbroek van Niemand", 100.00, 10.00, "Nintendo", "Ja")
KledingTest5 = Poloshirt("Poloshirt van Polo", 67.67, 67.67, "OldJeans", "Nee")

print(KledingTest1.beschrijving())
print(KledingTest2.beschrijving())
print(KledingTest3.beschrijving())
print(KledingTest4.beschrijving())
# Op ene andere manier is het beetje afgebroken, wel toont het de goede informatie. Geen idee waarom. Komt door de console I guess.
print(KledingTest5.beschrijving())

```

## DataFrames uitsplitsen en opbouwen met Series

De volgende 5 kolommen hebben betrekking op de contactdetails van elke medewerker in dit DataFrame:
- SALES_STAFF_CODE
- WORK_PHONE
- EXTENSION
- FAX
- EMAIL

Maak van elk van deze 5 kolommen een serie.

```python
con = sqlite3.connect('./go_sales.sqlite')

sql_query = "SELECT SALES_STAFF_CODE, WORK_PHONE, EXTENSION, FAX, EMAIL FROM sales_staff"
data = pd.read_sql_query(sql_query, con)

data_ssc = data['SALES_STAFF_CODE']
data_wp = data['WORK_PHONE']
data_ex = data['EXTENSION']
data_f = data['FAX']
data_em = data['EMAIL']

# Check the datatypes
print(type(data)) # dataframe (entiteit; hele tabel)
print(type(data_ssc)) # series (attribuut; een waarde van een tabel)
```

## Data wijzigen

### Datatypen

Door het attribuut .dtypes van een DataFrame op te vragen krijg je een serie die per kolom het datatype weergeeft. Doe dit bij de originele databasetabel *sales_staff*

```python
con = sqlite3.Connection('./go_sales.sqlite')

sales_staff = pd.read_sql_query("SELECT * FROM sales_staff", con)
sales_staff
```

Hier valt op dat elke kolom het datatype 'object' heeft: Python leest dus alles als string. Wiskundige operaties zijn hierdoor niet mogelijk.<br>
We kunnen proberen om kolommen met getallen, bijvoorbeeld de 'extension', te converteren naar een int. Zie onderstaande code.

```python
sales_staff['EXTENSION'] = sales_staff['EXTENSION'].astype(int)
sales_staff.dtypes
```

Dit lukt echter niet, omdat er in de kolom 'EXTENSION' lege waarden zitten die niet geconverteerd kunnen worden naar een int.<br>
Wél kan je deze naar een float converteren, zie onderstaande code:

```python
sales_staff['EXTENSION'] = sales_staff['EXTENSION'].astype(float)
sales_staff.dtypes
```

Als we in de rij van 'EXTENSION' kijken zien we dat de conversie van het datatype nu is gelukt.<br>
Dit is <b>randvoorwaardelijk</b> voor het uitvoeren van wiskundige operaties.<br>
