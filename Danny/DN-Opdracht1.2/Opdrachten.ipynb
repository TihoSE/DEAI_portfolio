Geef een overzicht met daarin van de producten behorend tot het producttype ‘Cooking Gear’ per productnaam de totaal verkochte hoeveelheid en de gemiddelde verkoopprijs. Sorteer de uitkomst op totaal verkochte hoeveelheid. (6 kolommen, 10 rijen)

```python
order_details = pd.read_sql_query("SELECT * FROM order_details", con)
product_query = pd.read_sql_query("SELECT * FROM product", con)
product_type_query = pd.read_sql_query("SELECT * FROM product_type", con)

product_all = pd.merge(product_query, product_type_query, how='inner', on='PRODUCT_TYPE_CODE')

cooking_gear = product_all[product_all['PRODUCT_TYPE_EN'] == 'Cooking Gear']

merge_all = pd.merge(cooking_gear, order_details, how='inner', on='PRODUCT_NUMBER')

resultaat = (
    merge_all
    .groupby(['PRODUCT_NUMBER','PRODUCT_NAME','PRODUCT_TYPE_CODE','PRODUCT_TYPE_EN'])
    # agg is average
    .agg(totaal_verkocht=('QUANTITY','sum'),gemiddelde_verkoopprijs=('UNIT_SALE_PRICE','mean')).reset_index()
    # sort_values werkt net als ORDER BY + waar je een naam kunt geven
    .sort_values(by='totaal_verkocht', ascending=False)
)

# head werkt net als de 'TOP' zoals SELECT TOP 10 
resultaat = resultaat.head(10)

resultaat
```

Geef een overzicht met daarin per (achternaam van) medewerker of hij/zij manager is of niet, door deze informatie toe te voegen als extra 'Ja/Nee'-kolom.<br>
Hint: gebruik een for-loop waarin je o.a. bepaalt of het woord 'Manager' in de functie (position_en) staat. (2 kolommen, 102 rijen).

```python
staff_query = pd.read_sql_query("SELECT * FROM sales_staff", con)

resultaat_lijst = []

# iterrows() zorgt ervoor dat elke rij wordt gesplitst in een indexnummer en een row, waarin alle rijwaarden opgeslagen liggen.
# Ik kan nog beetje herinneren dat bij het veld 'index' dat ik '_' ook kan gebruiken. De '_' betekend dat je het niet hoeft te gebruiken en niet kan gebruiken.
for index, row in staff_query.iterrows():
    if 'Manager' in row['POSITION_EN']:
        manager = 'Ja'
    else:
        manager = 'Nee'
    
    # List toegepast :), append zorgt ervoor dat de waardes toegevoegd wordt.
    # Heb dictionary in een list toegepast
    resultaat_lijst.append({'LAST_NAME': row['LAST_NAME'],'Manager': manager})

# Heb ook Dataframe toegepast
resultaat = pd.DataFrame(resultaat_lijst)

resultaat
```

```python
staff_query = pd.read_sql_query("SELECT * FROM sales_staff", con)

staff_query

# Roept de huidige jaar op
huidige_jaar = date.today().year

resultaat_lijst = []

for _, row in staff_query.iterrows():
    # Datum omzetten naar datetime
    hire_date = datetime.strptime(row['DATE_HIRED'], '%Y-%m-%d')
    
    # Aantal jaren in dienst (trekt van de huidige jaar de hire_date af)
    jaren_in_dienst = huidige_jaar - hire_date.year

    # Classificatie
    if jaren_in_dienst < 30:
        dienstverband = 'kort in dienst'
    else:
        dienstverband = 'lang in dienst'

    resultaat_lijst.append(
        # Dictionary toevoegen in de list.
        {
        'LAST_NAME': row['LAST_NAME'],
        'DIENSTVERBAND': dienstverband
        }
    )

# Schilt weer code te typen.
resultaat = pd.DataFrame(resultaat_lijst)

resultaat
```
